# 3.FacetGrid plot

# 4.Boxplot or Whisker plot

```python
fig=plt.gcf()
fig.set_size_inches(10,7)
fig=sns.boxplot(x='Species',y='PetalLengthCm',data = iris,order=['Iris-virginica','Iris-versicolor','Iris-setosa'],linewidth=2.5,orient='v',dodge=False,hue='Species')
plt.show()
```

```python
iris.boxplot(by='Species',figsize=(12,6))
plt.show()
```

# 5.Strip plot

```python
fig=plt.gcf()
fig.set_size_inches(10,7)
fig=sns.stripplot(x='Species',y='SepalLengthCm',data=iris,jitter=True,edgecolor='gray',size=8,palette='winter',orient='v')
plt.show()
```

```python
# 6.Combining Box and Strip Plots
fig = plt.gcf()
fig.set_size_inches(10,7)
fig=sns.boxplot(x='Species',y='SepalLengthCm',data=iris,hue='Species')
fig=sns.stripplot(x='Species',y='SepalLengthCm',data=iris,jitter=True,edgecolor='gray',hue='Species')
plt.show()
```

```python

plt.figure(figsize=(8, 6))

ax = sns.boxplot(x='Species', y='PetalLengthCm', data=iris,hue='Species')

# Stripplot on the SAME axes
sns.stripplot( x='Species',y='PetalLengthCm',data=iris,hue='Species',jitter=True,edgecolor='gray', linewidth=1,ax=ax)        # ðŸ‘‰ important: draw on same axis

# Colors for the 3 species (in x-axis order)
colors = ['green', 'red', 'yellow']

# Color each box safely
for box, color in zip(ax.artists, colors):
    box.set_facecolor(color)
    box.set_edgecolor('black')

plt.show()
```

# 14.KDE Plot

```python
sub = iris[iris['Species'] == 'Iris-setosa']

sns.kdeplot(data=sub, x='SepalLengthCm', y='SepalWidthCm', cmap="plasma", fill=True, thresh=0.25)
plt.title('Iris-setosa')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

# 16.Stacked Histogram

```python
iris['Species'] = iris['Species'].astype('category')
iris.head(2)
```

```python
list1=list()
mylabels=list()
for gen in iris.Species.cat.categories:
    list1.append(iris[iris.Species==gen].SepalLengthCm)
    mylabels.append(gen)

h=plt.hist(list1,bins=30,stacked=True,rwidth=1,label=mylabels)
plt.legend()
plt.show()
```
